## 1. ¿Cuál es el BACDistance promedio (en km)? Renombra la columna resultante como "avg_bac_distancia."

```python
sql = """
SELECT AVG("BACDistance (in Km)") as "avg_bac_distancia" FROM school_details
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

##2. ¿Cuál es el BACDistance promedio (en km) según nivel educativo del colegio?  Ordena los resultados de promedio mayor a promedio menor.


```python
sql = """
SELECT SchoolLevel, AVG("BACDistance (in Km)") as "avg_bac_distancia" 
FROM school_details
GROUP BY SchoolLevel
ORDER BY avg_bac_distancia DESC
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df
```

##3. Repite la consulta anterior, pero solo muestra los resultados donde la distancia promedio sea menor a 10 km.


```python
sql = """
SELECT SchoolLevel, AVG("BACDistance (in Km)") as "avg_bac_distancia" 
FROM school_details
GROUP BY SchoolLevel
HAVING AVG("BACDistance (in Km)") < 10
ORDER BY avg_bac_distancia DESC
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

##4. Une las dos tablas (detalles de colegios y detalles de instalaciones de los colegios). Comienza mostrando todas las columnas, pero limita las filas resultantes a solo 2.


```python
sql = """
SELECT s.*, 
f.IsLibraryAvailable, 
f.AvailableBooksCount, 
f.IsFireExtgAvailable, 
f.ExtinguisherCount, 
f.HOIRoomAvailable, 
f.IsStaffRoomAvailable, 
f.NoofStaffRoom, 
f.IsLockerForStaffAvailable, 
f.IsSportsStoreRoomAvailable
FROM school_details s LEFT JOIN school_facility_details f
ON s.EstablishmentCode = f.EstablishmentCode
LIMIT 2
"""
# Lee la consulta sql en pandas para ver los resultados
df=pd.read_sql_query(sql, engine)
df.head()
```

## 5. Ahora, solo selecciona colegios donde el nivel de este sea Escuela Primaria y que no tengan biblioteca.


```python
sql = """
SELECT s.*, 
f.IsLibraryAvailable, 
f.AvailableBooksCount, 
f.IsFireExtgAvailable, 
f.ExtinguisherCount, 
f.HOIRoomAvailable, 
f.IsStaffRoomAvailable, 
f.NoofStaffRoom, 
f.IsLockerForStaffAvailable, 
f.IsSportsStoreRoomAvailable
FROM school_details s LEFT JOIN school_facility_details f
ON s.EstablishmentCode = f.EstablishmentCode
WHERE SchoolLevel = 'PS' AND IsLibraryAvailable='Not Available'
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

## 6. Toma la consulta anterior, pero ordena los resultados en orden alfabético por nombre.

```python
sql = """
SELECT s.*, 
f.IsLibraryAvailable, 
f.AvailableBooksCount, 
f.IsFireExtgAvailable, 
f.ExtinguisherCount, 
f.HOIRoomAvailable, 
f.IsStaffRoomAvailable, 
f.NoofStaffRoom, 
f.IsLockerForStaffAvailable, 
f.IsSportsStoreRoomAvailable
FROM school_details s LEFT JOIN school_facility_details f
ON s.EstablishmentCode = f.EstablishmentCode
WHERE s.SchoolLevel = 'PS' AND IsLibraryAvailable='Not Available'
ORDER BY EstablishmentName
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```
