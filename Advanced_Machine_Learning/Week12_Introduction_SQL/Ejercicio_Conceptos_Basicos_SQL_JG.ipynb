## Importación de librerías y base de datos.

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Cargar datos
school = pd.read_csv('https://s3.amazonaws.com/General_V88/boomyeah2015/codingdojo/curriculum/content/chapter/School_Details.csv')
# Asegúrate de que estén cargados correctamente
school.head(2)
```

```python
school.info()
```

```python
# crear motor sqlite
engine = create_engine('sqlite:///school.db', echo=True)
# crear conexión con el motor
conn = engine.connect()
```

```python
# agregar marco de datos como una tabla en sqlite
sqlite_table = "school_details"
school.to_sql(sqlite_table, conn, if_exists='fail')
```

## 1. Llama las columnas NombreEstablecimiento, NivelEscolar y SitioWeb.


```python
sql = """
SELECT EstablishmentName, SchoolLevel, Website FROM school_details
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

##2. Muestra el NombreEstablecimiento y DistanciaDistrito (en km) de colegios en los que DistanciaDistrito (en km) es más de 100 km.0


```python
sql = """
SELECT EstablishmentName, "DistrictHQDistance (in Km)" FROM school_details
WHERE "DistrictHQDistance (in Km)">100
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head(11)
```

##3. Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios donde el nivel escolar es Junior High School.


```python
sql = """
SELECT EstablishmentName, SchoolLevel, "DistrictHQDistance (in Km)" FROM school_details
WHERE SchoolLevel="JHS"
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

##4. Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios que contengan las palabras “Junior High” en su nombre.


```python
sql = """
SELECT EstablishmentName, SchoolLevel, "DistrictHQDistance (in Km)" FROM school_details
WHERE EstablishmentName LIKE '%JUNIOR HIGH%'
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

##5. Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios cuyo nombre comience con la letra “C”. Ordena los resultados de forma alfabética por nombre de colegio  y limita los resultados de la consulta a 10 líneas.


```python
sql = """
SELECT EstablishmentName, SchoolLevel, "DistrictHQDistance (in Km)" FROM school_details
WHERE EstablishmentName LIKE 'C%'
ORDER BY EstablishmentName 
LIMIT 10
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

##6. Muestra los nombres de los colegios de áreas urbanas. Ordena los resultados en orden alfabético inverso.


```python
sql = """
SELECT EstablishmentName, LocatedInRuralOrUrban FROM school_details
WHERE LocatedInRuralOrUrban = 'Urban Area'
ORDER BY EstablishmentName DESC
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

##7. Repite la consulta anterior, pero renombra la columna que se muestra como “nombre”.

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName AS Nombre, LocatedInRuralOrUrban  FROM school_details
WHERE LocatedInRuralOrUrban = 'Urban Area'
ORDER BY EstablishmentName DESC
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```
